```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
images_shot_bgr = []
images_shot_rgb = []
for i in range(3):
  images_shot_bgr.append(cv2.imread('WIN_Pro' + str(i+1) + '.jpg', cv2.IMREAD_COLOR))
  images_shot_rgb.insert(i, images_shot_bgr[i][: , : , ::-1])
  plt.subplot(1, 3, i+1)
  plt.imshow(images_shot_rgb[i])

```

```python
# draw rectangle

images_rect_annt = images_shot_rgb.copy()
for i in range(3):
  images_rect_annt[i] = images_rect_annt[i].astype(np.uint8)

cv2.rectangle(images_rect_annt[0], (130,150), (380,380), (0,0,255), thickness=5, lineType=cv2.LINE_8)
plt.subplot(1,3,1)
plt.imshow(images_rect_annt[0])

cv2.rectangle(images_rect_annt[1], (200,100), (470,400), (0,0,255), thickness=5, lineType=cv2.LINE_8)
plt.subplot(1,3,2)
plt.imshow(images_rect_annt[1])

cv2.rectangle(images_rect_annt[2], (200,100), (470,400), (0,0,255), thickness=5, lineType=cv2.LINE_8)
plt.subplot(1,3,3)
plt.imshow(images_rect_annt[2])
```

```python
def get_kernel(image):
  h, w = image.shape[:2]
  kernel_width = (w // 7) | 1
  kernel_height = (h // 7) | 1
  return (kernel_width, kernel_height)
```

```python
faces = []


faces.insert(0, images_rect_annt[0][150:380, 130:380])
faces[0] =cv2.GaussianBlur(faces[0], get_kernel(images_shot_rgb[0]), 0)
images_rect_annt[0][150:380, 130:380] = faces[0]

plt.imshow(images_rect_annt[0])
```

```python

faces.insert(1, images_rect_annt[1][100:400, 200:470])
faces[1] =cv2.GaussianBlur(faces[1], get_kernel(images_shot_rgb[1]), 0)
images_rect_annt[1][100:400, 200:470] = faces[1]

plt.imshow(images_rect_annt[1])
```
